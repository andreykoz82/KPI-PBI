```python
import pandas as pd
```

```python
%%time
sales = pd.read_excel("sales_extend.xlsx")
geo = pd.read_excel("geo.xlsx")
```

```python
month_mapping = {'Январь': 1, 'Февраль': 2, 'Март': 3, 'Апрель': 4, 'Май': 5, 'Июнь': 6, 'Июль': 7, 
             'Август': 8, 'Сентябрь': 9, 'Октябрь': 10, 'Ноябрь': 11, 'Декабрь': 12,}
month = sales['По месяцам'].str.replace(' г.', "").str.split(' ', expand=True)[0].map(month_mapping)
year = sales['По месяцам'].str.replace(' г.', "").str.split(' ', expand=True)[1]
df = month.to_frame().join(year)
df[0] = df[0].astype('str')
df[1] = df[1].astype('str')
df['date'] = df[0]+ '-' + df[1]
df['date'] = pd.to_datetime(df['date'])
```

```python
sales['По месяцам'] = df['date']
```

```python
sales = sales.merge(geo, left_on='Контрагент', right_on='Наименование').drop(['Наименование', 
                                                                              'Артикул', 'ИНН', 'Индекс',
                                                                             'Тип'], 1)
```

```python
num = sales._get_numeric_data()
num[num < 0] = 0
```

```python
sales = sales.set_index('По месяцам')
```

```python
sales.head()
```

```python
 geo_df = sales.groupby([pd.Grouper(freq='M'), 'Регион', 'Широта', 'Долгота'])['Количество'].sum().reset_index()
```

```python
geo_df['Year'] = geo_df['По месяцам'].dt.year
```

```python
import plotly.express as px

fig = px.scatter_geo(geo_df,
                     lat='Широта',
                     lon='Долгота',
                     color="Регион",
                     hover_name="Регион", 
                     size="Количество",
                     animation_frame="Year",
                     projection="natural earth")
fig.update_layout(width=1000, height=800)
fig.show()
```
